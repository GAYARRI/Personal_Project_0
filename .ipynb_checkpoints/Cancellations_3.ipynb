```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler,Normalizer
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import NearMiss
from collections import Counter
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
import xgboost as xgb
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.metrics import roc_curve, auc
# Importar bibliotecas necesarias
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
warnings.filterwarnings("ignore")

```

```python
file_path ='C:/Users/gayar/Personal_Project/Machine_Learning/Datos_Entreno.csv'  # Update with actual path
df = pd.read_csv(file_path)
```

```python
# initial pre-process
# Drop the 'id' column if present

if 'id' in df.columns:
    df = df.drop(columns=['id'])

# Encode categorical features
categorical_cols = ['segment', 'deposit', 'cust_type']
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Store encoder for future use

# Define features and target
X = df.drop(columns=['is_cancelled'])
y = df['is_cancelled']



numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Crear el boxplot para las variables numricas
plt.figure(figsize=(12, 8))
df[numerical_cols].boxplot(rot=45)  # Rotar etiquetas para mejor visualizacin
plt.title("Boxplot de Variables Cuantitativas")
plt.ylabel("Valores")
plt.xlabel("Variables")
plt.show()
```

```python
# Calcular el rango intercuartlico (IQR) para la variable 'per_day_price'
X2=X.copy()
Q1 = X2['per_day_price'].quantile(0.25)
Q3 = X2['per_day_price'].quantile(0.75)
IQR = Q3 - Q1

# Definir el lmite superior como Q3 + 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR


# Reemplazar valores que excedan el lmite superior con el propio lmite
X2['per_day_price'] = X2['per_day_price'].apply(lambda x: upper_limit if x > upper_limit else x)

# Verificar los cambios aplicados
print(X2['per_day_price'].describe())


# Calcular el rango intercuartlico (IQR) para la variable 'duration'
Q1_duration = X2['duration'].quantile(0.25)
Q3_duration = X2['duration'].quantile(0.75)
IQR_duration = Q3_duration - Q1_duration

# Definir el lmite inferior como Q1 - 1.5 * IQR
lower_limit_duration = Q1_duration - 1.5 * IQR_duration

# Reemplazar valores que estn por debajo del lmite inferior con el propio lmite
X2['duration'] = X2['duration'].apply(lambda x: lower_limit_duration if x < lower_limit_duration else x)

# Verificar los cambios aplicados
print(X2['duration'].describe())






# Seleccionar columnas numricas excluyendo la variable objetivo
numerical_cols = X2.select_dtypes(include=['int64', 'float64']).columns

# Crear el boxplot para las variables numricas
plt.figure(figsize=(12, 8))
X2[numerical_cols].boxplot(rot=45)  # Rotar etiquetas para mejor visualizacin
plt.title("Boxplot de Variables Cuantitativas")
plt.ylabel("Valores")
plt.xlabel("Variables")
plt.show()
```

```python


###################################################################   MODULO PARA DETERMIMAR LOS TIPOS DE VARIABLES QUE SE VAN A MANEJAR EN EL DATA FRAME   ################################################################################################################


working_df=X2.copy()

print(working_df.columns)


##############################################################################################################################################################################################################################################################################


def find_boolean_columns(dataframe):

 boolean_columns = []
 for col in dataframe.columns:
     unique_values = dataframe[col].dropna().unique()

     if len(unique_values)==2:
                boolean_columns.append(col)
 return boolean_columns

boolean_columns = find_boolean_columns(working_df)

print(f' \n HAY {len(boolean_columns)} VARIABLES BIVARIANTES CON LOS SIQUIENTES NOMBRES :\n {boolean_columns}' )


#####################################################################      SELECCIONA COLUMNAS DE TIPO CATEGORICO ############################################################################################################################################################


categorical_columns = working_df.select_dtypes(include=['object']).drop(columns=boolean_columns, errors='ignore').columns
print(f'\n HAY {len(categorical_columns)} VARIABLES CATEGORICAS CON LOS SIGUIENTES NOMBRES :\n {categorical_columns.tolist()}' )



#####################################################################      SELECCIONA COLUMNAS DE TIPO FECHA ( PREVIAMENE IDENTIFICADAS ) ##############################################################################################################################




# Seleccionar columnas de tipo fecha
datetime_columns = working_df.select_dtypes(include=['datetime']).drop(columns=boolean_columns, errors='ignore').columns
print(f'\n HAY {len(datetime_columns)} VARIABLES DE FECHA CON LOS SIGUIENTES NOMBRES :\n {datetime_columns.tolist()}' )


#####################################################################      SELECCIONA COLUMNAS NUMRTICAS Y LAS SEGMENTA EN RATIOS , SOFT( DE VARIACION DE RANGO MENOR ) Y CUANTITARIVAS /( VARIACION DE RANGO MAYOR) ############################################################################################################################################################



numeric_columns = working_df.select_dtypes(include=['number']).drop(columns=boolean_columns, errors='ignore').columns

def desc_numeric_columns(dataframe,numeric_columns):

 ratio_columns = []
 soft_columns = []
 continuas_columns = []

 for col in numeric_columns:

     max_values = dataframe[col].max(skipna=True)
     min_values = dataframe[col].min(skipna=True)
     RG=max_values-min_values

     set_values = dataframe[col].dropna().unique()


     if max_values <= 1.0 and min_values >= 0.0  and len(set_values )> 2.0 :

                ratio_columns.append(col)

     elif RG < 100 and RG >= 1 and len(set_values )> 2 :

                continuas_columns.append(col)

     else:
                soft_columns.append(col)

 return ratio_columns,continuas_columns,soft_columns


ratio_columns,soft_columns,continuas_columns = desc_numeric_columns(working_df,numeric_columns)





print(f' \n HAY {len(soft_columns)} SOFT_COLUMNS CON LOS SIQUIENTES NOMBRES :\n {soft_columns}' )
print(f' \n HAY {len(ratio_columns)} RATIOS CON LOS SIQUIENTES NOMBRES :\n {ratio_columns}' )
print(f' \n HAY {len(continuas_columns)} VARIABLES CONTINUAS CON LOS SIQUIENTES NOMBRES :\n {continuas_columns}')
```

```python
copyB_df=working_df.copy()

from IPython.core.interactiveshell import InteractiveShell
from sklearn.preprocessing import MinMaxScaler,OrdinalEncoder


InteractiveShell.est_node_interactivity='all'

scaler = MinMaxScaler()  # Inicializar el escalador


scaler_numeric = MinMaxScaler()

copyB_df[numeric_columns] = scaler.fit_transform(copyB_df[numeric_columns])



ordinal_encoder = OrdinalEncoder()



encoded_categorical = pd.DataFrame(ordinal_encoder.fit_transform(copyB_df[categorical_columns]), columns=categorical_columns)



encoded_categorical_df = pd.DataFrame(encoded_categorical, columns=categorical_columns)



normalized_df = pd.concat([copyB_df[numeric_columns], encoded_categorical_df] ,axis=1)




print(normalized_df)
```

```python



data_df = normalized_df.copy()




# Aplicar PCA
pca = PCA(n_components=4)
principal_components = pca.fit_transform(data_df)



# Crear un DataFrame con los componentes principales
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3','PC4'])



# Mostrar la varianza explicada por cada componente principal
explained_variance = pca.explained_variance_ratio_
print("Varianza explicada por cada componente principal:")
for i, variance in enumerate(explained_variance, 1):
    print(f"PC{i}: {variance:.2%}")

# Visualizar los componentes principales en un grfico 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(principal_df['PC1'], principal_df['PC2'], principal_df['PC3'], c='r', marker='o')

ax.set_xlabel('Componente Principal 1')
ax.set_ylabel('Componente Principal 2')
ax.set_zlabel('Componente Principal 3')
ax.set_title('PCA con 4 Componentes Principales')

plt.show()


loadings = pca.components_.T

# Crear un DataFrame con las cargas
loadings_df = pd.DataFrame(data=loadings, columns=['PC1','PC2','PC3','PC4'], index=normalized_df.columns)

# Imprimir las cargas de los componentes principales
print("Cargas de los Componentes Principales:")
print(loadings_df)

# Visualizar las cargas usando un grfico de calor
plt.figure(figsize=(8, 5))
sns.heatmap(loadings_df, annot=True, cmap='coolwarm', center=0)
plt.title('Cargas de los Componentes Principales')
plt.show()
```

```python


# Aplicar PCA y conservar solo los componentes ms informativos
pca = PCA()
X_pca_df = pd.DataFrame(principal_df, columns=[f'PC{i+1}' for i in range(data_df.shape[1])])
X_pca_sel=X_pca_df[['PC1','PC2','PC3','PC4']]

```

```python
# ENTRENAMIENTO PARA ENCONTRAR PARAMETROS OPTIMOS
# podemos correr el modelo sobre x_pca_sel  bin sobre data_df
# data_df es el conjunto de variables explicativas original normalizado 
# x_pca_sel son las 4 primeras componentes principales ( que reunen el 60% de la nube original)

# ====== Divisin del dataset ======
X_train, X_val, y_train, y_val = train_test_split(data_df, y, test_size=0.3, random_state=42, stratify=y)

# Divisin del conjunto de entrenamiento en entrenamiento y prueba
X_train_train, X_train_test, y_train_train, y_train_test = train_test_split(X_train, y_train, test_size=0.3, random_state=42, stratify=y_train)

# ====== Aplicar SMOTE solo en el conjunto de entrenamiento ======
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train_train, y_train_train)

#near_miss = NearMiss(version=3)  # Hay distintas versiones, ver explicacin abajo
#X_resampled, y_resampled = near_miss.fit_resample(X_train_train, y_train_train)




print(f"Distribucin de clases antes de Balanced_Sampling: {Counter(y_train_train)}")
print(f"Distribucin de clases despus de Balanced_Sampling: {Counter(y_resampled)}")

# ====== Definir el modelo XGBoost ======
xgb_clf = xgb.XGBClassifier(eval_metric='auc')

# ====== Definir los hiperparmetros a evaluar en GridSearchCV ======

param_grid_opt = {
    'n_estimators': [500],  # Aade valor medio
    'learning_rate': [0.05, 0.75,],  # Valores intermedios razonables, elimina 0.3 por posible sobreajuste
    'max_depth': [8],  # Ms rango (menos y ms profundidad)
    'subsample': [0.6, 0.75],  # Aade valor medio
    'colsample_bytree': [0.8, 0.9],  # Centrado en valores ms robustos (quita 0.3)
    'reg_alpha': [0.01, 0.1],  # Aade 0 (sin regularizacin L1) y 1 (mayor penalizacin)
    'reg_lambda': [4,5],  # Aade valor alto (5) para mayor regularizacin L2
    'min_child_weight': [1,2],  # Aade valor medio (3)
    'gamma': [0.1, 0.2],  # Aade 0 (sin restriccin) y ajusta valores medios
    'max_delta_step': [0, 0.5]  # Aade 3 para mejorar estabilidad en clasificacin desbalanceada
}


# ====== GridSearchCV para optimizacin ======
grid_search = GridSearchCV(
    estimator=xgb_clf,
    param_grid=param_grid_opt,
    cv=5,  # Validacin cruzada
    scoring='roc_auc',  # Optimiza AUC-ROC en lugar de accuracy
    n_jobs=-1,
    verbose=2
)

# ====== Entrenar el modelo con los datos balanceados ======
grid_search.fit(X_resampled, y_resampled)

# Obtener los mejores parmetros
best_params = grid_search.best_params_
print(f'Mejor combinacin de hiperparmetros: {best_params}')

# Obtener el mejor clasificador entrenado
best_xgb_clf = grid_search.best_estimator_

# ====== Evaluacin en el conjunto de prueba ======
y_train_test_pred = best_xgb_clf.predict(X_train_test)

print("Reporte de clasificacin en el conjunto de prueba:")
print(classification_report(y_train_test, y_train_test_pred))

print("Matriz de confusin en el conjunto de prueba:")
print(confusion_matrix(y_train_test, y_train_test_pred))

# ====== Evaluacin en el conjunto de validacin (sin SMOTE) ======
y_val_pred = best_xgb_clf.predict(X_val)

print("Reporte de clasificacin en el conjunto de validacin:")
print(classification_report(y_val, y_val_pred))

print("Matriz de confusin en el conjunto de validacin:")
print(confusion_matrix(y_val, y_val_pred))
```

```python

# EJECUCION GBOOST CON PARAMETROS OPTIMIZADOS

X_train, X_val, y_train, y_val = train_test_split(data_df, y, test_size=0.3, random_state=42, stratify=y)

# Divisin del conjunto de entrenamiento en entrenamiento y prueba
X_train_train, X_train_test, y_train_train, y_train_test = train_test_split(X_train, y_train, test_size=0.3, random_state=42, stratify=y_train)

# ====== Aplicar SMOTE solo en el conjunto de entrenamiento ======
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train_train, y_train_train)

#near_miss = NearMiss(version=1)  # Hay distintas versiones, ver explicacin abajo
#X_resampled, y_resampled = near_miss.fit_resample(X_train_train, y_train_train)




xgb_clf = xgb.XGBClassifier(eval_metric='auc')

# ====== Definir los hiperparmetros a evaluar en GridSearchCV ======

param_grid_opt = {
    'n_estimators': [500],  # Aade valor medio
    'learning_rate': [0.05],  # Valores intermedios razonables, elimina 0.3 por posible sobreajuste
    'max_depth': [7],  # Ms rango (menos y ms profundidad)
    'subsample': [0.5],  # Aade valor medio
    'colsample_bytree': [0.8],  # Centrado en valores ms robustos (quita 0.3)
    'reg_alpha': [0.03],  # Aade 0 (sin regularizacin L1) y 1 (mayor penalizacin)
    'reg_lambda': [4],  # Aade valor alto (5) para mayor regularizacin L2
    'min_child_weight': [1],  # Aade valor medio (3)
    'gamma': [0.1],  # Aade 0 (sin restriccin) y ajusta valores medios
    'max_delta_step': [0]  # Aade 3 para mejorar estabilidad en clasificacin desbalanceada
}


# ====== GridSearchCV para optimizacin ======
grid_search = GridSearchCV(
    estimator=xgb_clf,
    param_grid=param_grid_opt,
    cv=5,  # Validacin cruzada
    scoring='roc_auc',  # Optimiza AUC-ROC en lugar de accuracy
    n_jobs=-1,
    verbose=2
)

# ====== Entrenar el modelo con los datos balanceados ======
grid_search.fit(X_resampled, y_resampled)

# Obtener los mejores parmetros
best_params = grid_search.best_params_
print(f'Mejor combinacin de hiperparmetros: {best_params}')

# Obtener el mejor clasificador entrenado
best_xgb_clf = grid_search.best_estimator_

# ====== Evaluacin en el conjunto de prueba ======
y_train_test_pred1 = best_xgb_clf.predict(X_train_test)

print("Reporte de clasificacin en el conjunto de prueba:")
print(classification_report(y_train_test, y_train_test_pred1))

print("Matriz de confusin en el conjunto de prueba:")
print(confusion_matrix(y_train_test, y_train_test_pred1))

# ====== Evaluacin en el conjunto de validacin (sin SMOTE) ======
y_val_pred1 = best_xgb_clf.predict(X_val)

print("Reporte de clasificacin en el conjunto de validacin:")
print(classification_report(y_val, y_val_pred1))

print("Matriz de confusin en el conjunto de validacin:")
print(confusion_matrix(y_val, y_val_pred1))
```

```python
# VISUALIZACION XGB METHOD

# Calculate the probabilities for positive class
y_train_test_probs1 = best_xgb_clf.predict_proba(X_train_test)[:, 1]
y_val_probs1 = best_xgb_clf.predict_proba(X_val)[:, 1]

# Compute ROC curve and AUC for training set
fpr_train1, tpr_train1, _ = roc_curve(y_train_test, y_train_test_probs1)
roc_auc_train1 = auc(fpr_train1, tpr_train1)

# Compute ROC curve and AUC for validation set
fpr_val1, tpr_val1, _ = roc_curve(y_val, y_val_probs1)
roc_auc_val1 = auc(fpr_val1, tpr_val1)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_train1, tpr_train1, color='red', lw=2, label=f'Train AUC = {roc_auc_train1:.3f}')
plt.plot(fpr_val1, tpr_val1, color='blue', lw=2, label=f'Validation AUC = {roc_auc_val1:.3f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for XGB method ')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
